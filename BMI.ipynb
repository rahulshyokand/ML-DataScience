```python
%matplotlib inline
import numpy
import pandas
import matplotlib.pyplot as plot
from pylab import * 

data = pandas.read_excel('BMI.xlsx')
```

```python
weight = data[['Weight']]
height = data[['Height']]
age = data[['Age']]
waist = data[['Waist']]
bmi = data[['BMI']]

```

```python
from pylab import *
import matplotlib.pyplot as plot

axis= plot.axis()

axis = plot.xlabel('weight')
plot.ylabel('bmi')
scatter( weight , bmi )

```

```python
#removed nan dat point from the data set
trainWeight = weight[1:80]
testWeight = weight[80:]

trainBmi = bmi[1:80]
testBmi = bmi[80:]
```

```python
scatter(trainWeight , trainBmi)
```

```python
scatter(testWeight , testBmi)
```

```python
#converting data frame to numpy array
trainWeight=trainWeight.values
trainBmi = trainBmi.values
```

```python
trainBmi= [i[0] for i in trainBmi]
trainWeight = [i[0] for i in trainWeight]
```

```python
from scipy import stats
x = numpy.array(trainWeight)
y = numpy.array(trainBmi)
```

```python
p4 = numpy.poly1d(numpy.polyfit(x, y, 2))
```

```python
#formatting graph 
xp = np.linspace(32, 100, 10)

plot.scatter(x, y)
plot.plot(xp, p4(xp), c='r')
plot.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainBmi), p4(np.array(trainWeight)))

print(r2)
```
